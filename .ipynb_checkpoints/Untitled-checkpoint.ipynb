```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("worldcup_data.csv")
```

```python
import seaborn as sns
```

```python
#line chart
plt.figure(figsize=(10,6))
plt.bar(df["Year"], df["Teams"], color='orange')
plt.title("Number of Teams in FIFA World Cups (1930–2022)")
plt.xlabel("Year")
plt.ylabel("Number of Teams")
plt.show()
```

```python
#  Histogram – Distribution of Goals Scored
plt.figure(figsize=(8,5))
plt.hist(df["Goals_Scored"], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Total Goals Scored (1930–2022)")
plt.xlabel("Goals Scored")
plt.ylabel("Frequency")
plt.show()
```

```python
#  Goals over time
plt.figure(figsize=(10,6))
plt.plot(df["Year"], df["Goals_Scored"], marker='o')
plt.title("Total Goals Scored in FIFA World Cups (1930–2022)")
plt.xlabel("Year")
plt.ylabel("Goals Scored")
plt.grid(True)
plt.show()
```

```python
#  Scatter Plot – Goals vs Matches
plt.figure(figsize=(8,6))
plt.scatter(df["Matches_Played"], df["Goals_Scored"], color='green')
plt.title("Goals vs Matches Played")
plt.xlabel("Matches Played")
plt.ylabel("Goals Scored")
plt.grid(True)
plt.show()
```

```python
#  Pie Chart – Number of Titles per Country
title_counts = df["Winner"].value_counts()
plt.figure(figsize=(7,7))
plt.pie(title_counts, labels=title_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("FIFA World Cup Titles by Country (1930–2022)")
plt.show()
```

```python
#  Box Plot – Spread of Goals Scored
plt.figure(figsize=(6,5))
plt.boxplot(df["Goals_Scored"], patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Boxplot of Goals Scored Across Tournaments")
plt.ylabel("Goals Scored")
plt.show()
```

```python
#  Heatmap – Correlation between numerical features
plt.figure(figsize=(6,5))
sns.heatmap(df[["Goals_Scored", "Matches_Played", "Teams"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap: Goals, Matches, Teams")
plt.show()
```

```python
# 🏁 FIFA World Cup Data Visualization — Conclusion

# 1️⃣ Histogram (Goals Scored Distribution)
# - Total goals scored per tournament generally increased over time.
# - Most tournaments recorded between 100–170 goals, showing consistent scoring trends.

# 2️⃣ Bar Chart (Teams Over Years)
# - Number of participating teams grew from 13 in 1930 to 32 since 1998.
# - Indicates global expansion and inclusivity of the tournament.

# 3️⃣ Scatter Plot (Goals vs Matches Played)
# - Positive correlation between matches and goals scored.
# - More matches lead to more goals — larger tournaments produce higher goal totals.

# 4️⃣ Pie Chart (Winners by Country)
# - Few countries dominate World Cup history (Brazil, Germany, Italy).
# - Recent winners (Spain, France, Argentina) show increasing competition and balance.

# 5️⃣ Box Plot (Goals Scored)
# - Median goals per tournament: around 140–150.
# - Outliers include very high-scoring tournaments like 1954 and 1998–2014.

# 6️⃣ Heatmap (Correlation Analysis)
# - Strong positive correlation (~0.9) between Goals_Scored and Matches_Played.
# - More teams => more matches => more goals.
```

```python
mean_goals = df["Goals_Scored"].mean()
mean_matches = df["Matches_Played"].mean()
mean_teams = df["Teams"].mean()
```

```python
# Print all results neatly
print("📈 FIFA World Cup Statistics Summary")
print("----------------------------------")
print(f"Mean Goals Scored     : {mean_goals:.2f}")
print(f"Mean Matches Played   : {mean_matches:.2f}")
print(f"Mean Teams Participated: {mean_teams:.2f}")
```
